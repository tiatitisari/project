```python
import numpy as np 
import pandas as pd 
df = pd.read_csv('product_sales.csv') # load the data into pandas 

# initial look data structure and information 
print(df.head())
print(df.info()) 

# check missing data and duplicate data
print(df.isna().sum())
print('total unique customer id is:' + str(df['customer_id'].nunique()) + ' while total row is:' +str(len(df))) 

# check numerical data 
print(df.describe()) 

# check categorical data 
print(df['sales_method'].unique()) 
print(df['state'].unique()) 
```

```python
# Data cleansing and Validating  
df = df[df['years_as_customer']<=39] # exclude data error or years_as_customer > 39 years 
df['sales_method'] = df['sales_method'].replace('em + call','Email + Call').apply(str.lower) 

# Fixing naming inconsistency 
# initially, Check spread of data for each method to see is there any big revenue gap between sales method. As consideration to fill missing values in revenue. 
print(df.groupby('sales_method')['revenue'].agg([np.mean,np.median,np.max,np.min, np.var])) 
sales_method_mean = df.groupby('sales_method')['revenue'].mean().to_dict() # map mean by sales method 
df['revenue']=round(df['revenue'].fillna(df['sales_method'].map(sales_method_mean)),2) # fill missing value by revenue for each sales method 
```

```python
# check outlier for numerical data 
def outlier(a): 
    iqr = np.quantile(df[a],0.75)-np.quantile(df[a],0.25)
    upper = 1.5*iqr+ np.quantile(df[a],0.75)
    lower = np.quantile(df[a],0.25) - (1.5*iqr)
    return print(str(a)+" >> lower <"+str(lower)+" upper >"+str(upper))
print(outlier('nb_sold'),outlier('revenue'),outlier('nb_site_visits'))
print(df[df['revenue']==238.320000])
```

# What does the spread of the revenue look like overall? And for each method?

```python
# What does the spread of the revenue look like overall?
revenue= pd.DataFrame(round(df['revenue'].describe(),2).reset_index())
revenue.columns=['Parameter','Revenue Values']
revenue_pivot = revenue.pivot_table(columns='Parameter',values='Revenue Values')
revenue_pivot['var'] = df['revenue'].var()
revenue_pivot['iqr'] = np.quantile(df['revenue'],0.75)-np.quantile(df['revenue'],0.25)
print(revenue_pivot)

# Revenue Growth 
df_revenue = df.groupby('week')['revenue'].sum().reset_index()
df_revenue['aft_revenue'] = df_revenue['revenue'].shift(1).fillna(0)
df_revenue['growth']= ((df_revenue['revenue']/df_revenue['aft_revenue'])-1) *100
df_revenue['growth'] = df_revenue['growth'].fillna(0)

fig,axs=plt.subplots(figsize=(20,10),ncols=2,nrows =2)

ax1= sns.lineplot(y='revenue', x='week', data=df,marker='o',ax = axs[0,0],errorbar=None)
ax2 = sns.barplot(y='revenue',x='week',estimator = 'sum',data=df,ax=axs[0,1])
ax3 = sns.histplot(x='revenue',bins=5,data=df,ax=axs[1,0])
ax_growth = sns.lineplot(y='growth', x='week', data=df_revenue,marker='o',ax = axs[1,1],errorbar=None)

ax2.bar_label(ax2.containers[0],fmt='%.f')

ax1.set_title('Average Revenue by Week (Figure 3)',size=14)
ax1.set_xlabel('Week')
ax1.set_ylabel('Average Revenue')

ax2.set_title('Total Revenue by Week (Figure 4)',size=14)
ax2.set_ylabel('Revenue')

ax3.set_title('Distribution of Revenue (Figure 5)', size=14)
ax3.set_xlabel('Revenue')
ax3.set_ylabel('Count')

ax_growth.set_title('Revenue Growth Week on Week (Figure 6)', size =14) 
ax_growth.set_xlim(1,6)

mean_label = round(df.groupby('week')['revenue'].mean(),2).to_dict()
growth_label = df_revenue[['growth']].to_dict()

for i,j in mean_label.items():
    axs[0,0].annotate(str(j), xy=(i, j),ha='center',va='top',size='medium')

for i,j in growth_label.items():
    for a,b in j.items(): 
        axs[1,1].annotate(str(round(b,2))+' %', xy=(a+1, b),ha='center',va='top',size='medium')
        
plt.show()
```

In general, revenue has increased overtime, even though there's a slight decrement in Week 3 compared to Week 2 (see Figure 3 and Figure 4). Call has the lowest performance in generating revenue, so we shouldn't use this method again. 

Email method is the second rank of method that generated higher revenue compared to call method even though still lower than email+call. To be more convincing whether there is any significant difference between email and email+call method, I ran statistical testing between these two group to determine if there is a statistically significant difference in the means (average) of a continuous variable (such as revenue) between two independent groups (the two marketing methods in your case). I eliminated call method because I considered it as irrelevant method based what we already saw in previous graphs. 

The null hypothesis (H0) for this test is that there is no significant difference between the means of the two groups, while the alternative hypothesis (Ha) is that there is a significant difference.

```python
import scipy.stats as stats
# Perform hypothesis testing (e.g., ANOVA)
f_statistic, p_value = stats.ttest_ind(
    df[df['sales_method'] == 'email + call']['revenue'],
    df[df['sales_method'] == 'email']['revenue'],
    # df[df['sales_method'] == 'email + call']['revenue']
)
if p_value < 0.05:
    print("There is a significant difference between the strategies.")
else:
    print("There is no significant difference between the strategies.")
```

Result of statistical testing shows there's a significant difference between those 2 strategies, so based on all information stated, the best method to use in order to generate more revenue is email + call.
